```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
df = pd.read_csv(r"https://drive.google.com/uc?export=download&id=1xxDtrZKfuWQfl-6KA9XEd_eatitNPnkB")
```

## Split Data

```python
X = df.drop(columns = ['price'])
y = df['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 51)
display(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
X_train
```

## Feature Scaling

```python
# Scaling is not necessary in decison Tree and random forest classification
sc = StandardScaler()
sc.fit(X_train)
X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

## Train Data


```python
lr = LinearRegression()
lr.fit(X_train, y_train)
# y_pred_lr = lr.predict(X_test)
lr.score(X_test, y_test)
```

```python
dt = DecisionTreeRegressor(criterion = 'mae')
dt.fit(X_train, y_train)
# y_pred_dt = dt.predict(X_test)
dt.score(X_test, y_test)
```

```python
# It will take time 

rf = RandomForestRegressor(criterion = 'mae', n_estimators=100)
rf.fit(X_train, y_train)
# y_pred_dt = dt.predict(X_test)
rf.score(X_test, y_test)
```
